#### concat()

```python
data_series1 = {'Column1' :pd.Series([10,10,10,10,10,10,10]),
               'Column2' :pd.Series([10,10,10,10,10,10,10]),
               'Column3' :pd.Series([10,10,33,44,55,66,77,])}
df1=pd.DataFrame(data_series1)
```

```python
data_series2 = {'Column1' :pd.Series([11,12,13,14,15,16,17]),
               'Column2' :pd.Series([22,23,24,25,26,27,28])}
df2=pd.DataFrame(data_series2)
```

```python
frame=[df1,df2]
res = pd.concat(frame)
```
